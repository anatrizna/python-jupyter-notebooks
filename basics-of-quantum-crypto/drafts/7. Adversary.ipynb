Same as in classical cryptography, quantum cryptography security proves starts with introducing adversary - Eva.

Our Eva has a powerful quantum computer with huge quantum memory to store qubits. She can intercept, store and resend qubits without revealing her presistance. She also can access classical channel Asja and Balvis uses and read all their messages.

Her only limits are:<ul>
    <li>She cannot violate laws of physics.
    <li>She cannot fake messages Asja and Balvis distribute during classical communication. We assume that classical channel is authenticated, such that if Eva will be able to tamper it after Asja and Balvis ended QKD session and generated the key - it is no more relevant.
</ul>

Let's recap our BB84 noise-free protocol flow:

1. Asja distributes quantum states $\ket{x}$<sup>A</sup><sub>Œ∏<sup>A</sup></sub>, Balvis measures states on a chosen basis $\ket{x}$<sup>B</sup><sub>Œ∏<sup>B</sup></sub>
3. Balvis sends a receipt to Asja about received states $i$ and measurement bases Œ∏<sup>B</sup>.
4. Asja sends to Balvis Œ∏<sup>A</sup>, they discard rounds where Œ∏<sup>A</sup>‚â†Œ∏<sup>B</sup>
5. Asja randomly choses $n$ bits of remaining string:
    - Asja tells Balvis which rounds are tested, they exchange $x$<sup>A</sup><sub>n</sub> and $x$<sup>B</sup><sub>n</sub>
    - Asja and Balvis calculate $ùõø$, if $ùõø$=0 they continue with final rounds, if $ùõø$‚â†0 they abort protocol.
6. Asja choses extractor seed and sends it to Balvis, they generate $k$<sub>A</sub> = $Ext(x$<sup>A</sup><sub>Remain</sub>, $y)$ and $k$<sub>B</sub> = $Ext(x$<sup>B</sup><sub>Remain</sub>, $y)$

<h3>Burdensome authenticated channel</h3>

You may wonder why do we need to keep classical authenticated channel between our legitimate users. Why don't we make BB84 to be fully quantum?

<h4>Man-in-the-Middle attack</h4>
If there is no authenticated classical channel, Eva can perform famous <b>Man-in-the-middle attack</b> (MITM). MITM is an attack where the attacker secretly alters the communications between two parties who believe that they are directly communicating with each other. 

Let's assume that there is no classical channel.

STEP 1: Asja-Eva

<li>Asja prepares random string (16 bits) and encode those in randomly picked bases  $Œ∏$<sub>A</sub>.
<li>Asja sends qubits to Balvis, but Eva intercepts those.
<li>Eva plays Balvis role, measures qubits in randomly picked bases.
<li>Eva then informs Asja about bases she used $Œ∏$<sub>E</sub>, encoded as $\ket{0}$ when she used X basis, or $\ket{1}$ when she used Z basis.
<li>Asja thinks that she received information from Balvis and sends back (to Eva) $Œ∏$<sub>A</sub> encoded in the same way.
<li>Asja and Eva removes from their key string bits, where they used different bases.

STEP 2: Eva-Balvis

Asja don't know about Eva presistance and believes that she has a secret key with Balvis. 
Eva starts playing Asjas' role to generate key with Balvis:


<li>Eva prepares random string and encode those in randomly picked bases  $Œ∏$<sub>E</sub>.
<li>Eva sends qubits to Balvis.
<li>Balvis measures qubits in randomly picked bases and sends to Eva $Œ∏$<sub>B</sub>, encoded as $\ket{0}$ when he used X basis, or $\ket{1}$ when he used Z basis.
<li>Eva sends to Balvis $Œ∏$<sub>E</sub> encoded in the same way.
<li>Eva and Balvis removes from their key string bits, where they used different bases.
       


STEP 3: Asja-Eva-Balvis

Now, Asja and Eva share key $k$<sub>AE</sub>, while Eva and Balvis share key $k$<sub>BE</sub>. 
Eva can read <b>all</b> messages they exchange by intercepting and decoding, or send fake messages by encoding using keys she share with legitimate users.

<h4>Intercept-resend attack</h4>

Classical post-processing stage helps legitimate users to ensure that they are communicating with each other and reduce possibility of MITM attack.
One more huge benefit from classical post-processing is QBER calculation.  

In BB84 Eva has no possibility to pretend she is one of the legitimate users and act as MITM. She also cannot tamper classical channel, because it is authenticated. But she can follow another hacking stategy and perform <b>intercept-resend attack</b> on qubits. 

STEP 1: Intercepting qubits

Eva intercepts qubit Asja sent to Balvis and measures it in randomly picked basis $Œ∏$<sub>E</sub>, because she has no information about $Œ∏$<sub>A</sub>.

STEP 2: Preparing fake state

Eva cannot just intercept qubits, a big number of lost qubits will revel her presestance or will force users to switch to another channel (for example, if they consider that qubits are lost becuase the problems in the line). 
She needs to send 'fake' states for all rounds she intercepted, so Balvis will receive qubits, thinking those are from Asja.

But since $Œ∏$<sub>A</sub> is unknown to Eva, her best stategy is encode 'fake' states on $Œ∏$<sub>E</sub> based on her measurement result and send these qubits to Balvis.

For example:

<li>If Eva measured qubit on X basis and got result 1, then she will send state $\ket{1}$ to Balvis.
<li>If Eva measures qubit on Z basis and got 0, then she will prepare state $\ket{+}$ to Balvis.

<h3>Task 2: Results</h3>

Program intercept-resend attack on BB84 protocol:
The protocol is as follows:
<li>Asja creates 24-bit long random bit string, picks random bases, prepares qubits and sends those to Eva.
<li>Eva measures each qubit on randomly picked basis.
<li>Eva prepares 'fake' states and sends those to Balvis.
<li>Balvis measures each qubit on randomly picked basis.
<li>Sifting: Asja and Balvis compare basis, discard rounds where basis does not match.
<li>QBER: Asja and Balvis compare half of their key bits. Asja and Balvis randomly pick rounds and discard those from their final key.

```python
# import all necessary objects and methods for quantum circuits
from qiskit import QuantumRegister, ClassicalRegister, QuantumCircuit, execute, Aer
from random import randrange

#Source: awards/teach_me_qiskit_2018/cryptography/Cryptography.ipynb

def SendState(qc1, qc2, qc1_name):
    ''' This function takes the output of a circuit qc1 (made up only of x and 
        h gates and initializes another circuit qc2 with the same state
    ''' 
    
    # Quantum state is retrieved from qasm code of qc1
    qs = qc1.qasm().split(sep=';')[4:-1]

    # Process the code to get the instructions
    for index, instruction in enumerate(qs):
        qs[index] = instruction.lstrip()

    # Parse the instructions and apply to new circuit
    for instruction in qs:
        if instruction[0] == 'x':
            old_qr = int(instruction[5:-1])
            qc2.x(qreg[old_qr])
        elif instruction[0] == 'h':
            old_qr = int(instruction[5:-1])
            qc2.h(qreg[old_qr])
        elif instruction[0] == 'm': # exclude measuring:
            pass
        else:
            raise Exception('Unable to parse instruction')

def print_outcomes_in_reserve(counts): # takes a dictionary variable
    for outcome in counts: # for each key-value in dictionary
        reverse_outcome = ''
        for i in outcome: # each string can be considered as a list of characters
            reverse_outcome = i + reverse_outcome # each new symbol comes before the old symbol(s)
    return reverse_outcome

def split_list(bit_string, parts=1):
    length = len(bit_string)
    return [ bit_string[i*length // parts: (i+1)*length // parts] 
             for i in range(parts) ]
```

```python
#Solution

qreg = QuantumRegister(24)
creg = ClassicalRegister(24) 
asja = QuantumCircuit(qreg, creg, name='Asja')

send=[] #Initial bit string ot send
asja_basis=[]
balvis_basis=[] 
eva_basis=[]

#Creating random bit string
for i in range(24):
    bit = randrange(2)
    send.append(bit)
    
#Asja preparing qubits
for i, n in enumerate(send):
    if n==1:
        asja.x(qreg[i]) # apply x-gate

#Asja encoding
for i in range(24):
    r=randrange(2) 
    if r==0: #if bit is 0, then she encodes in X basis
        asja_basis.append('X')
    else: #if bit is 1, then she encodes in Z basis
        asja.h(qreg[i])
        asja_basis.append('Z')

eva = QuantumCircuit(qreg, creg, name='Eva') #Defining Eva circuit
SendState(asja, eva, 'Asja') #Asja sends states to Eva

#Eva intercepts and measures qubits
for i in range(24):
    r=randrange(2) #Randomly pick a basis
    if r==0: #if bit is 0, then measures in X basis
        eva.measure(qreg[i],creg[i])
        eva_basis.append('X')
    else: #if bit is 1, then measures in Z basis
        eva.h(qreg[i])
        eva.measure(qreg[i],creg[i])
        eva_basis.append('Z')

job = execute(eva,Aer.get_backend('qasm_simulator'),shots=1) #Eva has only has one shot to measure qubits
counts = job.result().get_counts(eva) # counts is a dictionary object in python
counts = print_outcomes_in_reserve(counts)

eva_received = list(map(int, counts)) #Saving Eva received string as a list

#Eva preparing fake states
eva_fake = QuantumCircuit(qreg, creg, name='Eva_fake')
for i, n in enumerate(eva_received): #if measured bit is 1 - apply X gate
    if n==1:
        eva_fake.x(qreg[i]) # apply x-gate

for i, n in enumerate(eva_basis):
    if i == 'Z': #if Eva used Z basis to measure qubit, she applies H gate for same round
        eva_fake.h(qreg[i])
    else:
        pass

balvis = QuantumCircuit(qreg, creg, name='Balvis') #Defining Balvis circuit
SendState(eva_fake, balvis, 'Eva') #Eva sends states to Balvis

#Balvis receives qubits
for i in range(24):
    r=randrange(2) #Randomly pick a basis
    if r==0: #if bit is 0, then measures in X basis
        balvis.measure(qreg[i],creg[i])
        balvis_basis.append('X')
    else: #if bit is 1, then measures in Z basis
        balvis.h(qreg[i])
        balvis.measure(qreg[i],creg[i])
        balvis_basis.append('Z')

job_b = execute(balvis,Aer.get_backend('qasm_simulator'),shots=1) #Balvis has only has one shot to measure qubits
counts_b = job_b.result().get_counts(balvis) # counts is a dictionary object in python
counts_b = print_outcomes_in_reserve(counts_b)

received = list(map(int, counts_b)) #Saving Balvis received string as a list
```

Now, our Asja and Balvis will proceed with phase 2 "Sifting" and exchange information about bases they used. 

```python
#Sifting
asja_key=[] #Asjas register for matching rounds
balvis_key=[] #Balvis register for matching rounds
for j in range(0,len(asja_basis)): #Going through list of bases 
    if asja_basis[j] == balvis_basis[j]: #Comparing
        asja_key.append(send[j])
        balvis_key.append(received[j]) #Keeping key bit if bases matched
    else:
        pass #Discard round if bases mismatched
```
